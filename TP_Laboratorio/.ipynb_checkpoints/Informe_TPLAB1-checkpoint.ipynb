# 2. Simulacion numérica en Python

Mediante las restricciones observadas en el prototipo pasabajos se realizo el diseño del mismo utilizando la aproximación de máxima planicidad.

```python
#%% Inicializacion de librerias
# Librerías externas NumPy, SciPy y Matplotlib
import scipy.signal as sig
import matplotlib.pyplot as plt
import numpy as np

# Librería de TC2
import pytc2.sistemas_lineales as tc2

#%% Inicializacion de graficos para el Notebook
plt.figure(1)
plt.close(1)

import matplotlib as mpl

fig_sz_x = 13
fig_sz_y = 7
fig_dpi = 80 # dpi
fig_font_size = 16

mpl.rcParams['figure.figsize'] = (fig_sz_x, fig_sz_y)
mpl.rcParams['figure.dpi'] = fig_dpi
plt.rcParams.update({'font.size':fig_font_size})
#%%  Fin - Inicialización de librerías
```

```python
#Defino los parametros de la plantilla
w0= 2*np.pi*300
wz= 2*np.pi*100

w0_n= w0/w0
wz_n= wz/w0

#Transformo a prototipo pasabajo
W0= 1/w0_n
Wz= 1/wz_n

alfa_max= 3 #en dB
ee2= 1

#El orden sale por observacion ya que al simular el prototipo pasabajo con el cero de
#transmision, la simulacion coincide con la restriccion de la consigna.
n= 3

print("- Aproximacion de Butterworth -")
print("Epsilon2= ",ee2)
print("Grado del filtro: ",n)


#Generacion del prototipo pasabajo

z,p,k= sig.buttap(n)

num_lp,den_lp= sig.zpk2tf(z,p,k)
tf_lp= sig.TransferFunction(num_lp,den_lp)

#Agrego cero de transmision
k= 1/(Wz/W0)**2 #Esta constante regula la ganancia en continua, para que la misma sea de 0dB
num_lp_notch= [1*k, 0, Wz**2 *k]
tf_lp_notch= sig.TransferFunction(num_lp_notch, den_lp)

tc2.analyze_sys(tf_lp, sys_name="Prototipo Pasabajo")
tc2.analyze_sys(tf_lp_notch, sys_name="Prototipo Pasabajo Notch")
```

Ahora, se tansforma el prototipo pasabajo **SIN** los ceros en la transferencia. El cero de transimision sera agregado luego en la transferencia pasa-alto:
- Transferencia prototipo pasabajo: $$\frac{\omega_0}{s+\omega_0} \cdot \frac{{\omega_0}^2}{s^2 + s\frac{\omega_0}{Q}+{\omega_0}^2}$$
- Transferencia pasa-alto: $$\frac{s}{s+\omega_0} \cdot \frac{s^2}{s^2 + s\frac{\omega_0}{Q}+{\omega_0}^2}$$

Con la transferencia transformada en pasa-alto, se agrega el cero de transmision:
$$\frac{s}{s+\omega_0} \cdot \frac{s^2 + {\omega_z}^2}{s^2 + s\frac{\omega_0}{Q}+{\omega_0}^2}$$

```python
#Transformacion a pasaalto
num_hp, den_hp= sig.lp2hp(num_lp, den_lp)
tf_hp= sig.TransferFunction(num_hp, den_hp)

k_hp= 1/(wz_n/w0_n)**2

#Agrego el cero de transmision
num_hp[2]= wz_n**2

tf_hp_notch= sig.TransferFunction(num_hp, den_hp)
sos_hp_notch = tc2.tf2sos_analog(num_hp, den_hp)
print("\nTransferencia Pasa-alto con el cero de transmision:")
tc2.pretty_print_SOS(sos_hp_notch)

tc2.analyze_sys(tf_hp, sys_name="Filtro Pasa-alto")
tc2.analyze_sys(tf_hp_notch, sys_name="Filtro Pasa-alto Notch")
```

Con la transferencia del filtro conseguida puede realizarse la implementación circuital.

# 5. Comparación entre estructuras


- En la primera estructura propuesta, la salida por el nodo Vo determina una transferencia fija del tipo pasa-alto notch. Puede ajustarse el valor de $\omega_0$ , Q y k ajustando el valor de los componentes.
- Por otro lado, la estructura propuesta por Schaumann permite obtener distintas transferencias como por ejemplo pasabajos, pasa-alto, elimina banda, entre otras detalladas en la seccion 4, mediante el agregado de ceros de transmisión arbitrarios.
- La estructura de Schaumann provee la facilidad de modificar el numerador de la transferencia mediante los coeficientes a, b y c para lograr las distintas transferencias. Estos coeficientes indican el porcentaje en que los componentes pasivos son levantados de masa, con 1 indicando un levantamiento total.
- El primer circuito propuesto no cuenta con ganancia en altas frecuencias, si o si será de 0dB. El de Schaumann permite ajustar esta ganancia con el término $2a-c$.
