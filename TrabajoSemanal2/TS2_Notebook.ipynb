# 2. Valores de los componentes

### 2.1 Conseguir $\omega_0 = 1$ y $Q = 3$

Observando la normalizacion obtenida para el circuito, si se desea obtener estos valores de $\omega_0$ y $Q$ los componentes deberán tomar los siguientes valores:  

- $R_1= \frac{\Omega_Z}{K}$
- $R_2 = 3\cdot\Omega_Z$
- $R_3 = \Omega_Z$
- $C= \frac{1}{\Omega_Z}$


### 2.2 Conseguir ganancia de 20dB

Para conseguir 20dB en la banda de paso, debo reemplazar $K = 10$ ya que $20\cdot log(10) = 20dB$. Entonces, sólo se modifica el valor de $R_1$:

- $R_1= \frac{\Omega_Z}{10}$

# 4. Simulación circuital en Python

```python
#%% Inicializacion de librerias
# Librerías externas NumPy, SciPy y Matplotlib
from scipy.signal import TransferFunction
import matplotlib.pyplot as plt
import numpy as np

# Librería de TC2
from pytc2.sistemas_lineales import pzmap, GroupDelay, bodePlot
#%%
```

```python
#%% Inicializacion de graficos para el Notebook
plt.figure(1)
plt.close(1)

import matplotlib as mpl

fig_sz_x = 13
fig_sz_y = 7
fig_dpi = 80 # dpi
fig_font_size = 16

mpl.rcParams['figure.figsize'] = (fig_sz_x, fig_sz_y)
mpl.rcParams['figure.dpi'] = fig_dpi
plt.rcParams.update({'font.size':fig_font_size})
#%%  Fin - Inicialización de librerías
```

```python
#%% Defino los parametros
w0=10**0 ; q=3 ; r=1 ; k=1

#%% Defino los componentes y la transferencia
R1= r/k; R2=q*r*w0; R3=r; C=1/(r*w0);

my_tf = TransferFunction( [k * w0**2], [1, w0/q, w0**2] )

#%% Gráficos
plt.close('all')

bodePlot(my_tf, fig_id=1, filter_description = 'Pasabajo')

pzmap(my_tf, fig_id=2, filter_description = 'Pasabajo') #S plane pole/zero plot

GroupDelay(my_tf, fig_id=3, filter_description = 'Pasabajo')
```
